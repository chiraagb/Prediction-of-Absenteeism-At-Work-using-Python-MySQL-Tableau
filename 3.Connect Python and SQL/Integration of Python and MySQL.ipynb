```python
df_new_obs = model.predicted_outputs()
# df_new_obs = df_new_obs.astype(int)
df_new_obs.head(10)
```

```python
insert_query = 'INSERT INTO predicted_outputs VALUES ('
for i in range (df_new_obs.shape[0]):
    values = []

    for j in range (df_new_obs.shape[1]):
        col_val = df_new_obs[df_new_obs.columns.values[j]][i]
        if col_val == True or col_val==False:
            values.append(str(int(col_val)))
        else:
            values.append(str(col_val))
    insert_query += ','.join(["%s"]*len(values)) % tuple(values)
    insert_query += '), ('
    
insert_query = insert_query[:-3]
        
```

```python
insert_query
```

```python
insert_query = insert_query[:-1] + ');'
insert_query
```

```python
cursor.execute(insert_query)
```
