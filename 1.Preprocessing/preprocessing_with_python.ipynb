# Final Checkpoint:


```python
df_preprocessed = df_reason_date_mod.copy()
df_preprocessed.head(10)
```

```python
bool_cols = ['Reason_1','Reason_2','Reason_3','Reason_4']
```

```python
df_preprocessed[bool_cols] = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
cat_cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
for col in cat_cols:
    codes, _ = pd.factorize(df_preprocessed[col])
    df_preprocessed[col] = codes
```

```python
df_preprocessed[bool_cols] = df_preprocessed[bool_cols].astype(int)
```
